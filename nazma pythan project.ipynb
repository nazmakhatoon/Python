```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("HR Analytics Data 1.csv")
data
```

```python
# print the first 5 rows of the DataFrame
data.head(5)
```

```python
# Display the summary statistics of numerical columns.
df=data.describe()
df
```

```python
#Check for missing values in each column

data.isna().sum()
```

```python
# Calculate the average 'MonthlyRent' for each'Department'

data.groupby("Department")['DailyRate'].mean()
```

```python
#Create a bar plot showing the count of employees in each 'JobRole'

sns.set(rc={'figure.figsize':(8,5)})
sns.set(style="white")
sns.set_context("poster",font_scale=0.5) #paper,notebook,talk,poster

ax=sns.countplot(x="Department",data=data,palette='CMRmap',hue="Department",legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Create a bar plot representing the distribution of 'MaritalStatus' among employees

sns.set(rc={'figure.figsize':(8,5)})
sns.set(style="white")
sns.set_context("poster",font_scale=0.5) #paper,notebook,talk,poster

ax=sns.countplot(x="MaritalStatus",data=data,palette='brg',hue="MaritalStatus" ,legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Use a boxplot to visualize the relationship of 'MonthlyRent'for each 'Gender'

ax=sns.boxplot(x="MonthlyRate",data=data, y="Gender",hue='Gender')
```

```python
#Create a bar plot to visualize the relayionship btween'Age' and 'YearsAtCompany'

sns.set(rc={'figure.figsize':(14,5)})
sns.barplot(x="Age",y="YearsAtCompany",data=data,palette='viridis',hue="Age",legend=False)
```

```python
#USE a bar plot to explore relationships between numerical colums('Age','DailyRate','MonthlyRate')

sns.set(rc={'figure.figsize':(14,5)})
sns.set(style="white")
plt.subplot(2,1,1)
sns.barplot(x="Age",y="DailyRate",data=data,palette='spring',hue='Age',legend=False)

sns.set(rc={'figure.figsize':(14,5)})
sns.set(style="white")
plt.subplot(2,1,2)
sns.barplot(x="Age",y="MonthlyRate",data=data,palette='ocean',hue='Age',legend=False)
```

```python
#Explore the distribution of 'EducationField' using a count plot

sns.set(rc={'figure.figsize':(10,5)})
sns.set(style="white")
sns.set_context("poster",font_scale=0.5) #paper,notebook,talk,poster

ax=sns.countplot(x="EducationField",data=data,palette='Dark2',hue="EducationField",legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Create a stacked bar plot to show the distribution of 'Attrition'within each 'Department'

sns.barplot(x="Department",y="Age",data=data,hue="Attrition",palette='twilight')
```

```python
#Create a bar plot to compare'JobSatisfaction',across different'JobRoles'


sns.set_context("poster",font_scale=0.3)#paper,notebook,talk,poster

sns.barplot(x="JobRole",y="JobSatisfaction",data=data,palette='viridis',hue="JobRole",legend=False)

```

```python
#Create a histogram to visualize the distribution of 'Age'

sns.set(rc={'figure.figsize':(16,6)})
sns.set(style="white")
sns.set_context("poster",font_scale=0.5) #paper,notebook,talk,poster

sns.histplot(data["Age"],color='Blue')
plt.show()
```
